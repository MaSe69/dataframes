```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.barplot(x="day", y="total_bill", data=tips)
```

```python
plt.style.use("dark_background")
ax.ci = None
ax = sns.barplot(x="day", y="total_bill", data=tips, ci=None)
```

```python
xValues = ["A", "B", "C", "D"]
df = pd.DataFrame(np.random.rand(5,4)*1000, columns=xValues)
df
g = sns.barplot(x="A", y="B", data=df, ci=None)
#ax = sns.scatterplot(x="A", y="B", data=df, ci=None)
```

## Figures - Subplots and Saving

```python
fig = plt.figure(
    figsize=(9,11),
    facecolor=None,
)
fig.suptitle(
    "Figure Title",
    fontsize=24,
    fontweight=800,
    verticalalignment="top",
    y = 0.96,
)
ax = plt.subplot(2, 2, 1)
ax.set_title("Small Title", fontsize=18, fontweight=400, y=1.02)
g = sns.barplot(
    data=df,
    ax=ax,
    palette="Reds",
    ci=None,
)
### Does not work as in VSCode - Not so much sub-plot space
```
