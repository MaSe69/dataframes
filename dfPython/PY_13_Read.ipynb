## Converter

```python
def roundSales(x):
    x = round(float(x),-2)
    x = int(x)
    return x
def dayOfWeek(x):
    if x == "1":
        x = "Mo"
    elif x == "2":
        x = "Th"
    elif x == "3":
        x = "Th"        
    elif x == "4":
        x = "Th"
    elif x == "5":
        x = "Fr"
    elif x == "6":
        x = "Sa"        
    else:
        x = "So"
    return x

colsIwannaUse = ["Store", "Date", "Customers", "Sales", "DayOfWeek"]

df = pd.read_csv(fileTrain, parse_dates=["Date"], converters={"DayOfWeek": dayOfWeek, "Sales": roundSales}, dtype={"Store":'str',"Customers":'int64'}, usecols = colsIwannaUse, nrows=10000) 

df[2000:2010]
```
