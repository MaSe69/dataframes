## Converter

```python
df = pd.read_csv(filename, converters={"Cylinders": int()})
df.Cylinders = df.Cylinders + 7
df.head(3)
```

```python
def square(x):
    x = int(x)
    return x*x
def extract(x):
    return x[0:2]
#return 3*int(x)
df = pd.read_csv(filename, converters={"Cylinders": square, "Horsepower": extract})
df.head(3)
```

## Missing Values

## Dates

when reading, parse_dates=['column name']  is usually enough to magically parse into a date format

```python
filename = "../ZZ_Data/co2.csv"
df = pd.read_csv(filename, parse_dates=["Date"])
from datetime import timedelta, date
df["DatePlus"]= df.Date + timedelta(days=180)
df.head(10)
```
