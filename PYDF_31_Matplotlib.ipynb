# Matplotlib

## Matplotlib Native

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
arr = [9, 13, 50, 51]
plt.plot(arr)
plt.ylabel('My Y-Axis Values')
plt.xlabel('My X-Axis Values')
```

## Pandas Dataframes and Matplotlib

```python
values = [[2, 3], [4, 6]]
df = pd.DataFrame(values, columns=['Type A', 'Type B'], 
                   index=['Index 1', 'Index 2'])
df.plot(lw=2, colormap='jet', marker='o', markersize=5,
         title='Video streaming dropout by category')
plt.title("Default: Line Chart")         
plt.xlabel('xlabel')
plt.ylabel('ylabel')
```

```python
ax = df.plot.bar(lw=2, colormap='jet', title='Bar Chart')
ax.set_xlabel('xlabel')
ax.set_ylabel('xlabel')
```
